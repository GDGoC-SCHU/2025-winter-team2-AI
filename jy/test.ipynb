```python
import google.generativeai as genai
import json

# 모델 생성
model = genai.GenerativeModel('gemini-1.5-flash')

# 사용자 입력 받기
print("여행지 추천을 위해 아래 항목들을 입력해주세요:")
location = input("여행지 위치 (예: 제주도, 유럽): ")
activity_type = input("여행 스타일 (자유롭게 입력): ")  
best_season = input("추천할 여행 시기 (예: 봄, 여름, 가을, 겨울): ")

# AI에 메시지 전송
query = f"""
당신은 여행 추천 AI입니다.  
아래 JSON 형식으로 {location}에서 {activity_type}을(를) 즐길 수 있는 여행지 3곳을 추천하세요.

응답 형식:
{{
    "recommendations": [
        {{
            "destination_name": "여행지 이름 1",
            "location": "{location}",
            "type_of_activity": "{activity_type}",
            "best_season": "{best_season}"
        }},
        {{
            "destination_name": "여행지 이름 2",
            "location": "{location}",
            "type_of_activity": "{activity_type}",
            "best_season": "{best_season}"
        }},
        {{
            "destination_name": "여행지 이름 3",
            "location": "{location}",
            "type_of_activity": "{activity_type}",
            "best_season": "{best_season}"
        }}
    ]
}}

반드시 위 JSON 형식을 따르고, 정확한 여행지를 추천하세요.
"""

response = model.generate_content(query)

# 응답 처리
try:
    response_data = json.loads(response.text)  # JSON 변환 시도
    print(json.dumps(response_data, indent=4, ensure_ascii=False))  # 한글로 보기 좋게 출력
except json.JSONDecodeError:
    print("⚠️ 응답이 JSON 형식이 아닙니다. 원본 출력:")
    print(response.text)
```
